```python
import cProfile
import numpy as np
import scipy.special as spp
```

```python


cProfile.run('spp.kv(0, 0.5)')
```

```python
a = np.random.choice(20, 50000)

def kv1(b):
    return spp.kv(b, 0.25)

def kv2(a):
    inputs, indices = np.unique(a, return_inverse=True)
    outputs = spp.kv(inputs, 0.25)
    return outputs[indices]

    
```

```python
cProfile.run('kv1(a)')
```

```python
cProfile.run('kv2(a)')
```
